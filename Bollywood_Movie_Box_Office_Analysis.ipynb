```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/bollywood_movies.csv')
df.head()
```

```python
avg_rating = df.groupby('year')['rating'].mean()
avg_rating.plot(kind='line', marker='o', figsize=(10,6), title='Avg Movie Ratings by Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

```python
movie_count = df['year'].value_counts().sort_index()
movie_count.plot(kind='bar', figsize=(12,6), color='coral', title='Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.show()
```
